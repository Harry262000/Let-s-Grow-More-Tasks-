### Task 4 (Beginner ) :  Image to Pencil Sketch with Python

## Description:

   We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

```python
#import library 
import cv2
import matplotlib.pyplot as plt
```

```python
#read image 
image=cv2.imread("D:\Python and Machine Learning\FriendsPicture.jpg")
```

```python
#Showing image using openCV
cv2.imshow('original image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Displaying using Matplotlib
plt.imshow(image)
plt.axis(False)
plt.savefig('temp.png')
plt.show()
```

Convert BGR to RGB: Method 1

```python
plt.imshow(image[:,:,::-1])
plt.axis(False)
plt.savefig('temp.png')
plt.show()
```

Convert BGR to RGB: Method 2

```python
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.axis(False)
plt.savefig('temp.png')
plt.show()
```

Convert Image to a Pencil Sketch

```python
# Convert to Grey Image
grey_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Invert Image
invert_image=cv2.bitwise_not(grey_image)
#invert_img=255-grey_image

# Blur image
blur_image=cv2.GaussianBlur(invert_image, (111,111),0)

# Invert Blurred Image
invblur_image=cv2.bitwise_not(blur_image)
#invblur_image=255-blur_image

# Sketch Image
sketch_image=cv2.divide(grey_image,invblur_image, scale=256.0)

# Save Sketch 
cv2.imwrite('sketch.png', sketch_image)

# Display sketch
cv2.imshow('sketch image',sketch_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(14,8))

plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_image)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.savefig('temp1.png')
plt.show()
```
